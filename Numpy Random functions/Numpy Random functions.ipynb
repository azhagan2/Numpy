```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([1,2,3,4,5,6])
plt.show()
```

```python
# Distplot

sns.distplot([0,1,2,3,4,5,6], hist = False)
```

```python
from numpy import random as r
```

```python
# Normal Distribution

x=r.normal(loc=1, scale=2,size=(2,3))
sns.distplot(x)
```

```python
# Normal Distribution withoit hist

sns.distplot(r.normal(size=1000), hist=False)
plt.show()                                                    # Normal Distribution
```

```python
# Binomial Distribution

x = r.binomial(n=10, p=0.5, size=10)
print(x)
```

```python
sns.distplot(r.binomial(n=10, p=0.5, size=1000), hist=True, kde=True)
plt.show()
```

```python
sns.distplot(r.normal(loc=50,scale=5, size=1000),hist=False, label='normal')
sns.distplot(r.binomial(n=100, p=0.5, size=1000), hist=False, label = 'binomial')

plt.show()
```

```python
# Poisson Distribution

from numpy import random

x = random.poisson(lam=2, size=10)

print(x)
```

```python
sns.distplot(r.poisson(lam=2, size=1000), kde=True)
plt.show()
```

```python
sns.distplot(r.binomial(n=1000, p=0.01, size=1000),hist=False, label='binomial')
sns.distplot(r.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show()
```

```python
x=r.uniform(size=(2,3))
print(x)
```

```python
sns.distplot(r.uniform(size=1000), hist=False)
plt.show()
```

```python
x = random.logistic(loc=1, scale=2, size=(2, 3))

print(x)
```

```python
sns.distplot(r.logistic(loc=1, scale=2, size=(2,4)),hist=False)
```

```python
sns.distplot(r.logistic(loc=1, scale=2, size=1000),hist=False)
```

```python
sns.distplot(r.normal(scale=2, size=1000), hist=False,label='normal')
sns.distplot(r.logistic(size=1000), hist=False, label='logistic')
```
